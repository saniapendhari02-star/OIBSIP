# Missing Data Handling

I checked the percentage of missing values in each column before cleaning the data. This helped me decide whether to fill the missing values or remove the column.

```python

df["Age"] = df["Age"].fillna(df["Age"].median())

df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])


df.drop("Cabin", axis=1, inplace=True)
```

```python
df.isnull().sum()
```

## Handling Missing Values

I filled the missing values in the Age column using the median and the Embarked column using the most common value (mode). Since the Cabin column had too many missing values, I removed it from the dataset.

```python
df.to_csv("Titanic_Cleaned.csv", index=False)
```

## Duplicate Rows

```python
df["Sex"] = df["Sex"].str.lower()
df["Embarked"] = df["Embarked"].str.upper()
```

## Standardizing Text

I standardized the text values to keep the data consistent. The values in the **Sex** column were converted to lowercase, and the **Embarked** column was converted to uppercase.

```python
Q1 = df["Fare"].quantile(0.25)
Q3 = df["Fare"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df["Fare"] < lower_limit) | (df["Fare"] > upper_limit)]

outliers
```

## Outlier Detection

I used the IQR (Interquartile Range) method to detect outliers in the **Fare** column. The outliers were identified for analysis but were not removed because they may represent valid passenger fares.
